```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/wadefagen/datasets/master/gpa/uiuc-gpa-dataset.csv')
df
```

```python
df.to_csv('uiuc_gpa_dataset.csv', index=False)
```

```python
df['Course Name'] = df['Subject'] + df['Number'].astype(str)
df = df.drop(columns=["Subject", "Number"])
```

```python
grade_cols = ["A+", "A", "A-", "B+", "B", "B-", "C+", "C", "C-", "D+", "D", "D-", "F"]

df_combined = df.groupby("Course Name", as_index=False)[grade_cols].sum()

df_combined
```

```python
gpa_map = {
    "A+": 4.00,
    "A": 4.00,
    "A-": 3.67,
    "B+": 3.33,
    "B": 3.00,
    "B-": 2.67,
    "C+": 2.33,
    "C": 2.00,
    "C-": 1.67,
    "D+": 1.33,
    "D": 1.00,
    "D-": 0.67,
    "F": 0.00
}

# Calculate total GPA
df_combined['Total GPA'] = sum(df_combined[grade] * gpa_map[grade] for grade in grade_cols)

# Calculate total students
df_combined['Total Students'] = df_combined[grade_cols].sum(axis=1)

# Calculate average GPA
df_combined['Average GPA'] = df_combined['Total GPA'] / df_combined['Total Students']

df_combined
```

```python
selected_columns = ["Course Name", "Average GPA"]
result = df_combined[selected_columns]
result
```

```python
result.to_csv('uiuc_gpa_final.csv', index=False)
```
